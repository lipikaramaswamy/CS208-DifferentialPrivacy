```python
import pandas as pd
```

```python
marginal_latino = pums.latino.sum()/len(pums)
marginal_married = pums.married.sum()/len(pums)
marginal_disability = pums.disability.sum()/len(pums)
puma_1101 = pums[pums.puma == 1101]
marginal_engab = pums.englishability.sum()/len(pums)
marginal_puma_1101 = len(puma_1101)/len(pums)
```

```python
25766* marginal_latino * marginal_married * marginal_disability * marginal_puma_1101 * marginal_engab
```

```python
pums.groupby(['puma','sex', 'uscitizen', 'married', 'englishability']).agg({'state': 'count'})
```
